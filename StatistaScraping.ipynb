```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.options import Options
import time
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
name = "lvmh"
language ="0" #FR = 3 ; EN = 2 ; ALL = 0
premium = False

options = webdriver.ChromeOptions()
options.headless = True

driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
driver.get('https://fr.statista.com/recherche/?q=' + name + '&qKat=newSearchFilter&sortMethod=idrelevance&language=' + language +'&isRegionPref=-1&statistics=1&infos=1&accuracy=and&isoregion=0&isocountrySearch=&category=0&interval=0&archive=1')
# Scroll en bas de la page
# driver.execute_script("window.scrollTo(0,2000)")
title = driver.title
print(title)
time.sleep(2)
driver.find_element_by_id('onetrust-accept-btn-handler').click()
```

```python

try:
    main = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located(
            (By.XPATH, "//section[@class='padding-all-20  padding-horizontal-s-0']"))
    )
    
    artTitles = []
    artDesc = []
    artLink = []
    
    #Statistiques
    statArt = main.find_elements(
        By.XPATH, "//li[@class='list__item list__item--searchContentTypeStatistic pos-relative']")
    for item in statArt:
        
        statTitle = item.find_element_by_class_name(
            'h5.text-color--primaryDark.margin-bottom-0.text-ellipsis').text
        statDesc = item.find_element_by_class_name(
            'font-size-s.text-ellipsis').text
        
        link = item.find_elements_by_css_selector(
            "[href]")[0]
        medLink = link.get_attribute('href')
        
        artTitles.append(statTitle)
        artDesc.append(statDesc)
        artLink.append(medLink)
    
    #Infographies
    statArt = main.find_elements(
        By.XPATH, "//li[@class='list__item list__item--searchContentTypeInfographic pos-relative']")
    for item in statArt:

        statTitle = item.find_element_by_class_name(
            'h5.text-color--primaryDark.margin-bottom-0.text-ellipsis').text
        statDesc = item.find_element_by_class_name(
            'font-size-s.text-ellipsis').text
        link = item.find_elements_by_css_selector(
            "[href]")[0]
        medLink = link.get_attribute('href')
        
        artTitles.append(statTitle)
        artDesc.append(statDesc)
        artLink.append(medLink)
    
    
    #DataFrame
    df = pd.DataFrame({'Title': artTitles, 'Description': artDesc, 'Link': artLink})
        
except:
    print("Erreur au nivau du DataFrame")

finally:
    driver.quit()
    
```
